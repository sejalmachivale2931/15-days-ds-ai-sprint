```python
import pandas as pd
```

```python
df = pd.read_csv("train_and_test2.csv")
```

```python
df.head()
```

```python
print(df.columns.tolist())
```

```python
df = df[['Age', 'Fare', 'Sex', 'sibsp', 'Pclass', 'Embarked', '2urvived']]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
import numpy as np
age_mean = np.mean(df['Age'])
age_median = np.median(df['Age'])
age_std = np.std(df['Age'])

print("Mean :", age_mean)
print("Median :", age_median)
print("Standard Deviation :", age_std)
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR 
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Fare'] < lower_bound) | (df['Fare'] > upper_bound)]
print('Number Of Outliers : ',len(outliers))
print("Upper Bound : ", upper_bound)
outliers.head()
```

```python
df['Fare_log'] = np.log1p(df['Fare'])
df[['Fare','Fare_log']].head()
```

```python
df['Family_Size']= df['sibsp']+1
df[['sibsp', 'Family_Size']].head()
```
